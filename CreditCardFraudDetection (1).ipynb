```python
import pandas as pd

def find_outliers_using_iqr(data):
    bad_indx = []
    for col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        bad_nums = data[col][(data[col] < lower_bound) | (data[col] > upper_bound)]
        bad_indx.extend(bad_nums.index)

    return bad_indx

def remove_outliers(index, data):
    index = set(index)
    data = data.drop(index, axis=0)
    return data

# Example usage with a DataFrame df
bad_indx = find_outliers_using_iqr(df)
df_cleaned = remove_outliers(bad_indx, df)

df
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, confusion_matrix, recall_score
from sklearn.metrics import roc_auc_score, auc
```

```python
import numpy as np

# Convert DataFrames to NumPy arrays
X_train_array = X_train.values  # or X_train.to_numpy()
X_test_array = X_test.values     # or X_test.to_numpy()

# Reshape data to (samples, time_steps, features)
# Here we assume 1 time step for simplicity, you can adjust it based on your needs
X_train_reshaped = X_train_array.reshape((X_train_array.shape[0], 1, X_train_array.shape[1]))
X_test_reshaped = X_test_array.reshape((X_test_array.shape[0], 1, X_test_array.shape[1]))

# Check new shapes
print("X_train reshaped:", X_train_reshaped.shape)
print("X_test reshaped:", X_test_reshaped.shape)
```

```python
from keras.models import Model
from keras.layers import LSTM, Dense, Dropout, Attention, Input

# Define the input layer
inputs = Input(shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]))

# Add the first LSTM layer
x = LSTM(128, return_sequences=True)(inputs)  # Returns sequences for Attention
x = Dropout(0.2)(x)

# Split the LSTM output into query and value for the Attention layer
query = x
value = x

# Add the Attention layer
attention_output = Attention()([query, value])

# Add the second LSTM layer
x = LSTM(64, return_sequences=True)(attention_output)  # Returns sequences for Attention
x = Dropout(0.2)(x)

# Use a Dense layer to ensure dimensions match for the Attention layer
x = Dense(128)(x)  # Adjust the output dimension to match the value tensor
x = Dropout(0.2)(x)

# Add another Attention layer
x = Attention()([x, value])  # Use the output from the previous LSTM layer as the query
x = LSTM(32)(x)  # No return_sequences here
x = Dropout(0.2)(x)

# Add a Dense layer for the final output
x = Dense(16, activation='relu')(x)
x = Dropout(0.1)(x)
outputs = Dense(1, activation='sigmoid')(x)

# Define the model
model = Model(inputs=inputs, outputs=outputs)

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train_reshaped, y_train, validation_data=(X_test_reshaped, y_test), epochs=20, batch_size=32)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc, classification_report
from keras.models import Model
from keras.layers import LSTM, Dense, Dropout, Attention, Input
```

```python

# Make predictions
y_pred = (model.predict(X_test_reshaped) > 0.5).astype("int32")  # Convert probabilities to binary predictions
y_pred_proba = model.predict(X_test_reshaped)  # Get predicted probabilities for ROC curve

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

# Classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
import numpy as np
import itertools
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Function to plot confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion Matrix',
                          cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized Confusion Matrix")
    else:
        print("Confusion Matrix, without normalization")

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Confusion matrix
cm = confusion_matrix(y_test, predicted_classes)
plot_confusion_matrix(cm, classes=np.unique(y_test), title='Confusion Matrix')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, confusion_matrix
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define the number of folds
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize lists to store metrics for each fold
accuracy_scores = []
roc_aucs = []

# Loop over each fold
for fold, (train_index, test_index) in enumerate(skf.split(X_train, y_train), 1):
    print(f"Training fold {fold}...")

    # Split data into train and test for the current fold using iloc for integer-based indexing
    X_tr, X_te = X_train.iloc[train_index], X_train.iloc[test_index]
    y_tr, y_te = y_train.iloc[train_index], y_train.iloc[test_index]

    # Define the pipeline for the current fold
    pipeline_xg = Pipeline([
        ('smote', SMOTE(random_state=0)),
        ('xgb', XGBClassifier(random_state=5, eta=0.4))
    ])

    # Fit the pipeline to the training data of the current fold
    pipeline_xg.fit(X_tr, y_tr)

    # Get XGBoost predictions for train and test sets
    xgb_train_predictions = pipeline_xg.predict_proba(X_tr)[:, 1].reshape(-1, 1)
    xgb_test_predictions = pipeline_xg.predict_proba(X_te)[:, 1].reshape(-1, 1)

    # Scale the predictions
    scaler = StandardScaler()
    xgb_train_predictions_scaled = scaler.fit_transform(xgb_train_predictions)
    xgb_test_predictions_scaled = scaler.transform(xgb_test_predictions)

    # Define and compile the MLP model
    mlp_model = Sequential()
    mlp_model.add(Dense(32, activation='relu', input_shape=(xgb_train_predictions_scaled.shape[1],)))
    mlp_model.add(Dropout(0.2))
    mlp_model.add(Dense(16, activation='relu'))
    mlp_model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification
    mlp_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Set up callbacks for early stopping and learning rate reduction
    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=0.0001)
######################################################################
    # Train the MLP model on the current fold
    mlp_model.fit(
        xgb_train_predictions_scaled, y_tr,
        validation_data=(xgb_test_predictions_scaled, y_te),
        epochs=30, batch_size=32, verbose=0,
        callbacks=[early_stopping, reduce_lr]
    )

    # Evaluate the MLP model on the test data of the current fold
    loss, accuracy = mlp_model.evaluate(xgb_test_predictions_scaled, y_te, verbose=0)
    accuracy_scores.append(accuracy)

    # Get predictions and compute ROC AUC
    y_pred_prob = mlp_model.predict(xgb_test_predictions_scaled).ravel()
    roc_auc = roc_auc_score(y_te, y_pred_prob)
    roc_aucs.append(roc_auc)

    print(f"Fold {fold} - Accuracy: {accuracy:.4f}, ROC AUC: {roc_auc:.4f}")

# Calculate and print the mean accuracy and ROC AUC across all folds
mean_accuracy = np.mean(accuracy_scores)
mean_roc_auc = np.mean(roc_aucs)
print(f"\nMean Accuracy: {mean_accuracy:.4f}")
print(f"Mean ROC AUC: {mean_roc_auc:.4f}")

# After cross-validation, retrain on the full training data and evaluate on the original test set
pipeline_xg.fit(X_train, y_train)
xgb_train_predictions = pipeline_xg.predict_proba(X_train)[:, 1].reshape(-1, 1)
xgb_test_predictions = pipeline_xg.predict_proba(X_test)[:, 1].reshape(-1, 1)

# Scale the predictions
xgb_train_predictions_scaled = scaler.fit_transform(xgb_train_predictions)
xgb_test_predictions_scaled = scaler.transform(xgb_test_predictions)


```
